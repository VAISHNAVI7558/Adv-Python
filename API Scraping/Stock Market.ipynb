## Class creation

```python
import pandas as pd
import requests
import plotly.graph_objects as go
## create class
class API_STOCK_MARKET:
    def __init__(self  ):
        self.api_key = os. getenv('API_KEY')
        self.headers = {
            "x-rapidapi-key":api_key , 
            "x-rapidapi-host": "alpha-vantage.p.rapidapi.com"
        }
        self.url = "https://alpha-vantage.p.rapidapi.com/query"

        # EXTRACT SYMBOLS
    def extract_symbols(self , keywords):
        querystring = {"datatype":"json",
                       "keywords":keywords,
                       "function":"SYMBOL_SEARCH"}
        response = requests.get(self.url, headers=self.headers, params=querystring)
        data = response.json()
        dict1 =  {}
        for  i  in  data['bestMatches']:
            symbols  =  i['1. symbol']
            dict1[symbols] = [i['2. name'] ,  i['4. region'] , i['8. currency']]
        return dict1
    # FETCH DAILY TIME SERIES
    def fetch_daily_time_series(self , symbol):
        querystring = {"function":"TIME_SERIES_DAILY",
                       "symbol":symbol,
                       "outputsize":"compact",
                       "datatype":"json"}
        response = requests.get(self.url, headers=self.headers, params=querystring)
        data1  = response.json()
        # CONVERT TO DATAFRAME
        df1 = pd.DataFrame(data1['Time Series (Daily)'])
        df =  df1.T
        df.index = pd.to_datetime(df.index)
        # GIVE NAME TO INSEX
        df = df.astype(float)
        return df
    # PLOT CANDLESTICK CHART
    def plot (self , symbol):
        fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['1. open'],
                high=df['2. high'],
                low=df['3. low'],
                close=df['4. close'])])
        fig.update_layout(
            title=f'Candlestick chart for {symbol}',
            yaxis_title='Price',
            xaxis_title='Date',
            xaxis_rangeslider_visible=False
        )
        fig.show()


    
```
